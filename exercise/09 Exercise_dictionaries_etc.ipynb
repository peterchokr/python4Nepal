### Key Explanation Points

**Dictionary Fundamentals**
- Dictionaries use curly braces {} and store key-value pairs
- Keys must be unique and immutable (strings, numbers, tuples)
- Values can be any data type including lists, dictionaries, etc.
- Use square brackets [key] or get() method to access values

**Safe Dictionary Access**
- get() method returns None or default value if key doesn't exist
- Square brackets [] raise KeyError if key doesn't exist
- Always check key existence with 'in' operator when necessary
- Use get() for optional keys, [] for required keys

**Dictionary Methods**
- keys(): returns all keys in the dictionary
- values(): returns all values in the dictionary  
- items(): returns key-value pairs for iteration
- pop(): removes and returns value for given key

**Sets**
- Sets store only unique elements (no duplicates)
- Use curly braces {1, 2, 3} or set() function
- Common operations: intersection (&), union (|), difference (-)
- Perfect for removing duplicates from lists

**Data Structure Selection**
- Dictionary: for key-value mappings and fast lookups
- Set: for unique collections and set operations
- List: for ordered, mutable sequences
- Tuple: for immutable, ordered data
