- **Learning Objectives**: Understanding programming concepts and mastering Python development environment
- **학습목표**: 프로그래밍의 개념 이해 및 파이썬 개발환경 완전히 익히기
- **Content**:
  - What is programming? (프로그래밍이란 무엇인가?)
  - Features and applications of Python (파이썬의 특징과 활용 분야)
  - Python 3.x installation (Windows) (Python 3.x 설치)
  - Development environment introduction and installation (IDLE, VS Code) (개발환경 소개 및 설치)
  - Detailed IDLE usage (IDLE 사용법 상세 설명)
  - VS Code installation and Python extension setup (VS Code 설치 및 Python 확장 설정)
  - Code execution methods (Interactive mode vs Script mode) (코드 실행 방법)
- **Practice**:
  - (1) Python installation and first "Hello, World!" program (파이썬 설치 및 첫 번째 프로그램 작성)
  - (2) Running same code in IDLE and VS Code, comparing differences (IDLE과 VS Code에서 같은 코드 실행 및 차이점 비교)
  - (3) Creating simple self-introduction program (간단한 자기소개 프로그램 만들기)
- **Quiz**:
  - Q1: Write what appears on screen when print("안녕하세요") is executed (print("안녕하세요")를 실행했을 때 화면에 출력되는 내용을 쓰시오)
  - Q2: What symbols are used for comments in Python? Write examples of single-line and multi-line comments (파이썬에서 주석을 표시하는 기호는 무엇인가요? 한 줄 주석과 여러 줄 주석을 각각 작성해보세요)
  - Q3: Explain 2+ differences between IDLE and VS Code, describing pros and cons of each (IDLE과 VS Code의 차이점을 2가지 이상 설명하고, 각각의 장단점을 서술하시오)

## 💡 **Learning Tips**
**학습 팁**

1. **Weekly Practice-Centered**: 1 hour theory + 2 hours practice
   (매주 실습 중심: 이론 1시간 + 실습 2시간으로 구성)
2. **Step-by-Step Progression**: Understand previous week's content before proceeding
   (단계별 진행: 이전 주차 내용을 확실히 이해한 후 다음 단계로)
3. **Repetitive Learning**: Important concepts used repeatedly across multiple weeks
   (반복 학습: 중요한 개념은 여러 주에 걸쳐 반복 활용)
4. **Project-Based**: Experience solving real-life problems with Python
   (프로젝트 기반: 실생활 문제를 파이썬으로 해결하는 경험 제공)

## 🎯 **Expected Outcomes**
**기대 효과**

Upon completing this curriculum, students will be able to:
이 커리큘럼을 완주하면 학생들은:

- Master basic Python syntax completely (파이썬의 기본 문법을 완전히 익힐 수 있습니다)
- Understand fundamentals of object-oriented programming (객체지향 프로그래밍의 기초를 이해할 수 있습니다)
- Solve real-world problems through programming (실생활 문제를 프로그래밍으로 해결할 수 있는 능력을 갖추게 됩니다)
- Work with web APIs and data scraping (웹 API와 데이터 크롤링을 활용할 수 있습니다)
- Build foundation for learning other programming languages (다른 프로그래밍 언어를 학습할 수 있는 기반을 마련합니다)
