### Key Explanation Points

**Function Basics**
- def keyword is required to define functions
- Colon (:) must follow function definition line
- Function body must be indented
- Functions can be called multiple times for code reuse

**Parameters vs Arguments**
- Parameters: variables in function definition
- Arguments: actual values passed when calling function
- Default parameters must come after required parameters
- Keyword arguments use parameter names for clarity

**Return Values**
- return statement sends value back to caller
- Functions without return statement return None
- Returned values can be stored in variables or used in expressions
- print() shows output but doesn't return usable values

**Variable Scope**
- Local variables: created inside functions, only accessible within
- Global variables: created outside functions, accessible everywhere
- Use global keyword to modify global variables inside functions
- Local variables don't affect global variables with same name

**Function Design**
- Use descriptive function names
- Keep functions focused on single tasks
- Default parameters provide flexibility
- Functions calling other functions enable complex problem solving
