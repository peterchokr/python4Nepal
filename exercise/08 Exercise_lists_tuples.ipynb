### Key Explanation Points

**List Basics**
- Lists use square brackets [] and are zero-indexed
- Support negative indexing (-1 for last element)
- Mutable: can add, remove, and modify elements
- Allow duplicate values and mixed data types

**List Operations**
- append(): adds element to end
- insert(): adds element at specific position
- remove(): removes first occurrence of value
- pop(): removes and returns element (default: last)
- Slicing syntax: [start:end:step] where end is excluded

**Tuples**
- Use parentheses () and are immutable
- Single element tuple requires comma: (42,)
- More memory efficient than lists for fixed data
- Support unpacking: x, y = (10, 20)

**List Comprehensions**
- Syntax: [expression for item in iterable if condition]
- More concise than traditional loops for simple operations
- Can include filtering with conditional statements
- Should be kept readable - use regular loops for complex logic

**When to Choose**
- Lists: for data that changes, need methods like append/remove
- Tuples: for fixed data, coordinates, function returns
- Performance: tuples are faster for read-only operations
