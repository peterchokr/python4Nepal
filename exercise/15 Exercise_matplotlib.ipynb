### Key Explanation Points

**Basic Matplotlib Usage**
- Standard import: `import matplotlib.pyplot as plt`
- Always use `plt.show()` to display graphs
- `plt.figure()` creates new figure, `figsize` sets dimensions
- Proper labeling with `plt.title()`, `plt.xlabel()`, `plt.ylabel()`

**Graph Types and Functions**
- Line graphs: `plt.plot()` - best for trends over time
- Bar charts: `plt.bar()` for vertical, `plt.barh()` for horizontal
- Histograms: `plt.hist()` with `bins` parameter for data distribution
- Scatter plots: `plt.scatter()` for correlation analysis

**Styling and Enhancement**
- Colors: use color names or hex codes
- Markers: 'o', 's', '^', '*' for different point styles
- Transparency: `alpha` parameter (0.0 to 1.0)
- Grid: `plt.grid(True)` for better readability
- Legend: `plt.legend()` for multiple data series

**Advanced Features**
- Subplots: `plt.subplot(rows, cols, index)` for multiple graphs
- Annotations: `plt.annotate()` and `plt.text()` for labels
- Statistical lines: `plt.axhline()`, `plt.axvline()` for reference lines
- Layout: `plt.tight_layout()` for proper spacing

**Data Visualization Best Practices**
- Choose appropriate graph type for data nature
- Use clear, descriptive titles and labels
- Ensure color accessibility and contrast
- Avoid misleading scales or truncated axes
- Include legends and units where necessary
