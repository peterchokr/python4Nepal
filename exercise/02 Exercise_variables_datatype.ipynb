**Variable Fundamentals**
- Variables are labeled storage locations in memory
- Store, retrieve, and modify data throughout program execution
- Essential for making programs dynamic and reusable

**Naming Rules (Mandatory)**
- Start with letter or underscore (_)
- Use only letters, numbers, underscores
- Case-sensitive (name â‰  Name â‰  NAME)
- Cannot use Python keywords (if, for, class, etc.)

**Data Types Overview**
- **int**: Whole numbers (age = 20)
- **float**: Decimal numbers (height = 175.5)
- **str**: Text data ("John Kim")
- **bool**: True/False values (is_student = True)

**Best Practices**
- Use meaningful, descriptive names
- Follow snake_case convention (student_name)
- Be consistent throughout the project
- Use English variable names for international collaboration

**Type Function**
- `type(variable)` returns data type
- Useful for debugging and validation
- Returns class information (`<class 'int'>`)

**Common Mistakes**
- Starting names with numbers (2nd_name)
- Using hyphens or special characters
- Mixing naming conventions
- Confusing string numbers ("25") with integers (25)
