**String Fundamentals**
- Strings are sequences of characters in quotes
- Immutable (cannot be changed, only create new strings)
- Created with single ('') or double ("") quotes
- Support Unicode characters and special symbols

**Indexing and Slicing**
- Indexing starts at 0 (text[0] = first character)
- Negative indexing from end (text[-1] = last character)
- Slicing syntax: [start:end:step] (end not included)
- Advanced slicing: [::-1] reverses string

**Essential String Methods**
- Case conversion: .upper(), .lower(), .title(), .capitalize()
- Whitespace: .strip(), .lstrip(), .rstrip()
- Search/Replace: .find(), .count(), .replace()
- Content checking: .isdigit(), .isalpha(), .isalnum()

**String Formatting**
- f-strings: f"Hello {name}" (modern, preferred)
- .format() method: "Hello {}".format(name)
- Expressions in f-strings: f"{value:.2f}"
- Multiple variables: f"{first} {last}"

**Common Applications**
- Data cleaning and standardization
- Email and input validation
- Text analysis and processing
- Template generation and formatting
