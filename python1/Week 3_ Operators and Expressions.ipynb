### 1.1 Basic Arithmetic Operators (ê¸°ë³¸ ì‚°ìˆ  ì—°ì‚°ì)

## ğŸ”¢ 5. Operator Precedence (ì—°ì‚°ì ìš°ì„ ìˆœìœ„)

Understanding operator precedence helps you predict the order of operations.
ì—°ì‚°ì ìš°ì„ ìˆœìœ„ë¥¼ ì´í•´í•˜ë©´ ì—°ì‚° ìˆœì„œë¥¼ ì˜ˆì¸¡í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

### 5.1 Order of Operations (ì—°ì‚° ìˆœì„œ)

**Remember PEMDAS** (ê´„í˜¸, ì§€ìˆ˜, ê³±ì…ˆ/ë‚˜ëˆ—ì…ˆ, ë§ì…ˆ/ëº„ì…ˆ ìˆœì„œë¥¼ ê¸°ì–µí•˜ì„¸ìš”):

1. **P**arentheses (ê´„í˜¸) - `( )`
2. **E**xponents (ì§€ìˆ˜) - `**`
3. **M**ultiplication and **D**ivision (ê³±ì…ˆê³¼ ë‚˜ëˆ—ì…ˆ) - `*`, `/`, `//`, `%`
4. **A**ddition and **S**ubtraction (ë§ì…ˆê³¼ ëº„ì…ˆ) - `+`, `-`

### 5.2 Examples with Explanation (ì„¤ëª…ì´ ìˆëŠ” ì˜ˆì‹œ)

````
# Example 1: Basic precedence
result1 = 10 + 3 * 2
print(f"10 + 3 * 2 = {result1}")  # 16 (not 26!)
# Step by step: 3 * 2 = 6, then 10 + 6 = 16

# Example 2: Using parentheses
result2 = (10 + 3) * 2
print(f"(10 + 3) * 2 = {result2}")  # 26
# Step by step: (10 + 3) = 13, then 13 * 2 = 26

# Example 3: Complex expression
result3 = 2 ** 3 + 4 * 5 - 6 / 2
print(f"2 ** 3 + 4 * 5 - 6 / 2 = {result3}")  # 25.0
# Step by step: 2**3=8, 4*5=20, 6/2=3.0, then 8+20-3.0=25.0

# Use parentheses for clarity (ëª…í™•ì„±ì„ ìœ„í•´ ê´„í˜¸ ì‚¬ìš©)
price = 100
tax_rate = 0.1
total = price + (price * tax_rate)  # Clear what's being calculated
print(f"Total with tax: ${total}")  # $110.0
````
