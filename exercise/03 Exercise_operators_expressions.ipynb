### Key Explanation Points

**Arithmetic Operators**
- Basic: +, -, *, / (returns float)
- Advanced: // (floor division), % (modulo), ** (exponentiation)
- Division: / gives float result, // gives integer result

**Comparison Operators**
- Equality: == (equal), != (not equal)
- Ordering: <, >, <=, >= 
- Return boolean values (True/False)
- Works with numbers, strings, and other data types

**Logical Operators**
- and: Both conditions must be True
- or: At least one condition must be True  
- not: Reverses the boolean value
- Used to combine multiple conditions

**Assignment Operators**
- Compound: +=, -=, *=, /= (shortcuts for updating variables)
- More efficient than writing x = x + value
- Commonly used in loops and calculations

**Operator Precedence (PEMDAS)**
- Parentheses first
- Exponentiation (**)
- Multiplication/Division (*, /, //, %)
- Addition/Subtraction (+, -)
- Use parentheses for clarity

**Common Mistakes**
- Confusing = (assignment) with == (comparison)
- Forgetting operator precedence rules
- Not using parentheses for complex expressions
- Mixing up / and // operators
