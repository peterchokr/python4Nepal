## 🛠️ 2. Building Complete Programs (완전한 프로그램 구축)

### Common Program Patterns (일반적인 프로그램 패턴)

#### Pattern 1: Menu System (메뉴 시스템)

````python
while True:
    print("1. Option A")
    print("2. Option B") 
    print("3. Exit")
    
    choice = input("Choose: ")
    
    if choice == "1":
        print("You chose A")
    elif choice == "2":
        print("You chose B")
    elif choice == "3":
        break
    else:
        print("Invalid choice")
````

#### Pattern 2: Data Collection (데이터 수집)

````python
students = []

while True:
    name = input("Enter name (or 'done'): ")
    if name == "done":
        break
    
    grade = int(input("Enter grade: "))
    students.append({"name": name, "grade": grade})

print(f"Total students: {len(students)}")
````

#### Pattern 3: Data Processing (데이터 처리)

````python
# Calculate statistics
total = sum(student["grade"] for student in students)
average = total / len(students)
print(f"Average: {average:.1f}")

# Find high performers
for student in students:
    if student["grade"] >= 90:
        print(f"High performer: {student['name']}")
````

### Lab 1: Simple Grade Manager (간단한 성적 관리자)

**Problem (문제)**: Create a simple program to manage student grades with basic menu options.
기본 메뉴 옵션으로 학생 성적을 관리하는 간단한 프로그램을 만드세요.

**Requirements (요구사항)**:
- Add students with grades (성적과 함께 학생 추가)
- View all students (모든 학생 보기)
- Calculate class average (학급 평균 계산)

**Solution (정답)**:

````python
# Simple Grade Manager
students = []

print("SIMPLE GRADE MANAGER")
print("=" * 25)

while True:
    print("\n--- MENU ---")
    print("1. Add Student")
    print("2. View Students")
    print("3. Calculate Average")
    print("4. Exit")
    
    choice = input("Choose option (1-4): ")
    
    if choice == "1":
        # Add student
        print("\n--- ADD STUDENT ---")
        name = input("Enter student name: ")
        
        # Get grade with validation
        while True:
            try:
                grade = int(input("Enter grade (0-100): "))
                if 0 <= grade <= 100:
                    break
                else:
                    print("Grade must be 0-100!")
            except ValueError:
                print("Please enter a number!")
        
        students.append({"name": name, "grade": grade})
        print(f"Added {name} with grade {grade}")
    
    elif choice == "2":
        # View students
        print("\n--- ALL STUDENTS ---")
        if not students:
            print("No students found!")
        else:
            print("Name\t\tGrade")
            print("-" * 20)
            for student in students:
                print(f"{student['name']:<15} {student['grade']}")
    
    elif choice == "3":
        # Calculate average
        print("\n--- CLASS AVERAGE ---")
        if not students:
            print("No students found!")
        else:
            total = sum(student["grade"] for student in students)
            average = total / len(students)
            highest = max(student["grade"] for student in students)
            lowest = min(student["grade"] for student in students)
            
            print(f"Total students: {len(students)}")
            print(f"Average grade: {average:.1f}")
            print(f"Highest grade: {highest}")
            print(f"Lowest grade: {lowest}")
    
    elif choice == "4":
        print("Goodbye!")
        break
    
    else:
        print("Invalid choice! Please try again.")
````

### Lab 2: Simple Adventure Game (간단한 어드벤처 게임)

**Problem (문제)**: Create a simple choice-based adventure game where players make decisions that affect the story.
플레이어가 스토리에 영향을 주는 결정을 내리는 간단한 선택 기반 어드벤처 게임을 만드세요.

**Requirements (요구사항)**:
- Player makes choices (플레이어가 선택 결정)
- Different story paths (다른 스토리 경로)
- Simple scoring system (간단한 점수 시스템)

**Solution (정답)**:

````python
# Simple Adventure Game
import random

score = 0
player_name = input("Enter your character name: ")

print(f"\nWelcome to the adventure, {player_name}!")
print("Make choices to progress through the story.")

# Story Scene 1
print("\n--- SCENE 1: THE FOREST ---")
print("You are walking through a dark forest.")
print("You see two paths ahead.")
print("1. Take the left path (looks safe)")
print("2. Take the right path (looks mysterious)")

choice1 = input("What do you choose? (1 or 2): ")

if choice1 == "1":
    print("\nYou take the safe path and find a small village.")
    print("The villagers welcome you warmly.")
    score += 10
    path = "village"
elif choice1 == "2":
    print("\nYou take the mysterious path and find an old cave.")
    print("You hear strange sounds from inside.")
    score += 5
    path = "cave"
else:
    print("\nYou stand there confused. A wolf appears!")
    print("You run away quickly.")
    score += 0
    path = "forest"

# Story Scene 2
print(f"\n--- SCENE 2: THE {path.upper()} ---")

if path == "village":
    print("The village elder offers you a choice:")
    print("1. Help with farming (safe but slow)")
    print("2. Join the village guard (risky but rewarding)")
    
    choice2 = input("What do you choose? (1 or 2): ")
    
    if choice2 == "1":
        print("\nYou help with farming and earn the villagers' trust.")
        score += 15
    elif choice2 == "2":
        print("\nYou join the guard and help catch some bandits!")
        score += 25
    else:
        print("\nYou can't decide and miss both opportunities.")
        score += 5

elif path == "cave":
    print("Inside the cave, you find a treasure chest!")
    print("1. Open the chest immediately")
    print("2. Look around for traps first")
    
    choice2 = input("What do you choose? (1 or 2): ")
    
    if choice2 == "1":
        if random.random() < 0.5:  # 50% chance
            print("\nThe chest contains gold! You're rich!")
            score += 30
        else:
            print("\nIt's a trap! You lose some health but escape.")
            score += 5
    elif choice2 == "2":
        print("\nYou find and disable a trap, then safely get the treasure!")
        score += 20
    else:
        print("\nYou're too scared and leave empty-handed.")
        score += 0

else:  # forest path
    print("Lost in the forest, you find a magical fountain.")
    print("1. Drink from the fountain")
    print("2. Fill your water bottle and leave")
    
    choice2 = input("What do you choose? (1 or 2): ")
    
    if choice2 == "1":
        print("\nThe water gives you magical powers!")
        score += 20
    elif choice2 == "2":
        print("\nYou wisely save the water for later.")
        score += 10
    else:
        print("\nYou ignore the fountain and continue walking.")
        score += 5

# Final Results
print(f"\n--- ADVENTURE COMPLETE ---")
print(f"Congratulations, {player_name}!")
print(f"Your final score: {score}")

if score >= 30:
    print("🏆 Excellent! You're a true adventurer!")
elif score >= 20:
    print("👍 Good job! You made wise choices.")
elif score >= 10:
    print("😊 Not bad! You survived the adventure.")
else:
    print("🤔 Better luck next time!")

print("Thanks for playing!")
````

### Lab 3: Simple To-Do List (간단한 할일 목록)

**Problem (문제)**: Create a simple to-do list manager where users can add tasks, mark them complete, and view their progress.
사용자가 작업을 추가하고, 완료로 표시하고, 진행 상황을 볼 수 있는 간단한 할일 목록 관리자를 만드세요.

**Requirements (요구사항)**:
- Add new tasks (새 작업 추가)
- Mark tasks as complete (작업을 완료로 표시)
- View all tasks with status (상태와 함께 모든 작업 보기)

**Solution (정답)**:

````python
# Simple To-Do List Manager
tasks = []
task_id = 1

print("SIMPLE TO-DO LIST MANAGER")
print("=" * 30)

while True:
    print("\n--- MENU ---")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Complete Task")
    print("4. Show Statistics")
    print("5. Exit")
    
    choice = input("Choose option (1-5): ")
    
    if choice == "1":
        # Add task
        print("\n--- ADD TASK ---")
        title = input("Enter task title: ").strip()
        
        if title:
            task = {
                "id": task_id,
                "title": title,
                "completed": False
            }
            tasks.append(task)
            print(f"Added task: {title}")
            task_id += 1
        else:
            print("Task title cannot be empty!")
    
    elif choice == "2":
        # View tasks
        print("\n--- ALL TASKS ---")
        if not tasks:
            print("No tasks found!")
        else:
            print("ID\tStatus\t\tTask")
            print("-" * 35)
            for task in tasks:
                status = "✓ Done" if task["completed"] else "⏳ Pending"
                print(f"{task['id']}\t{status}\t\t{task['title']}")
    
    elif choice == "3":
        # Complete task
        print("\n--- COMPLETE TASK ---")
        if not tasks:
            print("No tasks found!")
        else:
            # Show pending tasks
            pending_tasks = [task for task in tasks if not task["completed"]]
            if not pending_tasks:
                print("No pending tasks!")
            else:
                print("Pending tasks:")
                for task in pending_tasks:
                    print(f"{task['id']}. {task['title']}")
                
                try:
                    task_id_input = int(input("Enter task ID to complete: "))
                    
                    # Find and complete task
                    task_found = False
                    for task in tasks:
                        if task["id"] == task_id_input:
                            if task["completed"]:
                                print("Task already completed!")
                            else:
                                task["completed"] = True
                                print(f"Completed: {task['title']}")
                            task_found = True
                            break
                    
                    if not task_found:
                        print("Task ID not found!")
                        
                except ValueError:
                    print("Please enter a valid task ID!")
    
    elif choice == "4":
        # Show statistics
        print("\n--- STATISTICS ---")
        if not tasks:
            print("No tasks found!")
        else:
            total_tasks = len(tasks)
            completed_tasks = len([task for task in tasks if task["completed"]])
            pending_tasks = total_tasks - completed_tasks
            completion_rate = (completed_tasks / total_tasks) * 100
            
            print(f"Total tasks: {total_tasks}")
            print(f"Completed: {completed_tasks}")
            print(f"Pending: {pending_tasks}")
            print(f"Completion rate: {completion_rate:.1f}%")
            
            if completion_rate >= 80:
                print("🎉 Great progress!")
            elif completion_rate >= 50:
                print("👍 Keep it up!")
            else:
                print("💪 Time to get busy!")
    
    elif choice == "5":
        print("Thanks for using To-Do List Manager!")
        break
    
    else:
        print("Invalid choice! Please try again.")
````

## 💡 Tips for Success (성공을 위한 팁)

### Programming Best Practices (프로그래밍 모범 사례)

1. **Plan before coding** (코딩 전에 계획)
   - Write down what your program should do (프로그램이 해야 할 일을 적어두기)
   - Break big problems into smaller steps (큰 문제를 작은 단계로 나누기)

2. **Use clear variable names** (명확한 변수명 사용)
   ```python
   # Good
   student_name = "Alice"
   final_grade = 85
   
   # Bad
   x = "Alice"
   y = 85
   ```

3. **Validate user input** (사용자 입력 검증)
   ```python
   while True:
       try:
           age = int(input("Enter age: "))
           if age > 0:
               break
           else:
               print("Age must be positive!")
       except ValueError:
           print("Please enter a number!")
   ```

### Common Mistakes to Avoid (피해야 할 일반적인 실수)

- Not handling invalid input (잘못된 입력 처리하지 않기)
- Forgetting to update loop variables (반복문 변수 업데이트 잊기)
- Using confusing variable names (혼란스러운 변수명 사용)
- Not testing edge cases (경계 사례 테스트하지 않기)
