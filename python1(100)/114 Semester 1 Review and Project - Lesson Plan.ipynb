### Lab 1: Simple Grade Manager (ê°„ë‹¨í•œ ì„±ì  ê´€ë¦¬ì)

**Problem (ë¬¸ì œ)**: Create a simple program to manage student grades with basic menu options.
ê¸°ë³¸ ë©”ë‰´ ì˜µì…˜ìœ¼ë¡œ í•™ìƒ ì„±ì ì„ ê´€ë¦¬í•˜ëŠ” ê°„ë‹¨í•œ í”„ë¡œê·¸ë¨ì„ ë§Œë“œì„¸ìš”.

**Requirements (ìš”êµ¬ì‚¬í•­)**:
- Add students with grades (ì„±ì ê³¼ í•¨ê»˜ í•™ìƒ ì¶”ê°€)
- View all students (ëª¨ë“  í•™ìƒ ë³´ê¸°)
- Calculate class average (í•™ê¸‰ í‰ê·  ê³„ì‚°)

**Solution (ì •ë‹µ)**:

````python
# Simple Grade Manager
students = []

print("SIMPLE GRADE MANAGER")
print("=" * 25)

while True:
    print("\n--- MENU ---")
    print("1. Add Student")
    print("2. View Students")
    print("3. Calculate Average")
    print("4. Exit")
    
    choice = input("Choose option (1-4): ")
    
    if choice == "1":
        # Add student
        print("\n--- ADD STUDENT ---")
        name = input("Enter student name: ")
        
        # Get grade with validation
        while True:
            try:
                grade = int(input("Enter grade (0-100): "))
                if 0 <= grade <= 100:
                    break
                else:
                    print("Grade must be 0-100!")
            except ValueError:
                print("Please enter a number!")
        
        students.append({"name": name, "grade": grade})
        print(f"Added {name} with grade {grade}")
    
    elif choice == "2":
        # View students
        print("\n--- ALL STUDENTS ---")
        if not students:
            print("No students found!")
        else:
            print("Name\t\tGrade")
            print("-" * 20)
            for student in students:
                print(f"{student['name']:<15} {student['grade']}")
    
    elif choice == "3":
        # Calculate average
        print("\n--- CLASS AVERAGE ---")
        if not students:
            print("No students found!")
        else:
            total = sum(student["grade"] for student in students)
            average = total / len(students)
            highest = max(student["grade"] for student in students)
            lowest = min(student["grade"] for student in students)
            
            print(f"Total students: {len(students)}")
            print(f"Average grade: {average:.1f}")
            print(f"Highest grade: {highest}")
            print(f"Lowest grade: {lowest}")
    
    elif choice == "4":
        print("Goodbye!")
        break
    
    else:
        print("Invalid choice! Please try again.")
````

### Lab 2: Simple Adventure Game (ê°„ë‹¨í•œ ì–´ë“œë²¤ì²˜ ê²Œì„)

**Problem (ë¬¸ì œ)**: Create a simple choice-based adventure game where players make decisions that affect the story.
í”Œë ˆì´ì–´ê°€ ìŠ¤í† ë¦¬ì— ì˜í–¥ì„ ì£¼ëŠ” ê²°ì •ì„ ë‚´ë¦¬ëŠ” ê°„ë‹¨í•œ ì„ íƒ ê¸°ë°˜ ì–´ë“œë²¤ì²˜ ê²Œì„ì„ ë§Œë“œì„¸ìš”.

**Requirements (ìš”êµ¬ì‚¬í•­)**:
- Player makes choices (í”Œë ˆì´ì–´ê°€ ì„ íƒ ê²°ì •)
- Different story paths (ë‹¤ë¥¸ ìŠ¤í† ë¦¬ ê²½ë¡œ)
- Simple scoring system (ê°„ë‹¨í•œ ì ìˆ˜ ì‹œìŠ¤í…œ)

**Solution (ì •ë‹µ)**:

````python
# Simple Adventure Game
import random

score = 0
player_name = input("Enter your character name: ")

print(f"\nWelcome to the adventure, {player_name}!")
print("Make choices to progress through the story.")

# Story Scene 1
print("\n--- SCENE 1: THE FOREST ---")
print("You are walking through a dark forest.")
print("You see two paths ahead.")
print("1. Take the left path (looks safe)")
print("2. Take the right path (looks mysterious)")

choice1 = input("What do you choose? (1 or 2): ")

if choice1 == "1":
    print("\nYou take the safe path and find a small village.")
    print("The villagers welcome you warmly.")
    score += 10
    path = "village"
elif choice1 == "2":
    print("\nYou take the mysterious path and find an old cave.")
    print("You hear strange sounds from inside.")
    score += 5
    path = "cave"
else:
    print("\nYou stand there confused. A wolf appears!")
    print("You run away quickly.")
    score += 0
    path = "forest"

# Story Scene 2
print(f"\n--- SCENE 2: THE {path.upper()} ---")

if path == "village":
    print("The village elder offers you a choice:")
    print("1. Help with farming (safe but slow)")
    print("2. Join the village guard (risky but rewarding)")
    
    choice2 = input("What do you choose? (1 or 2): ")
    
    if choice2 == "1":
        print("\nYou help with farming and earn the villagers' trust.")
        score += 15
    elif choice2 == "2":
        print("\nYou join the guard and help catch some bandits!")
        score += 25
    else:
        print("\nYou can't decide and miss both opportunities.")
        score += 5

elif path == "cave":
    print("Inside the cave, you find a treasure chest!")
    print("1. Open the chest immediately")
    print("2. Look around for traps first")
    
    choice2 = input("What do you choose? (1 or 2): ")
    
    if choice2 == "1":
        if random.random() < 0.5:  # 50% chance
            print("\nThe chest contains gold! You're rich!")
            score += 30
        else:
            print("\nIt's a trap! You lose some health but escape.")
            score += 5
    elif choice2 == "2":
        print("\nYou find and disable a trap, then safely get the treasure!")
        score += 20
    else:
        print("\nYou're too scared and leave empty-handed.")
        score += 0

else:  # forest path
    print("Lost in the forest, you find a magical fountain.")
    print("1. Drink from the fountain")
    print("2. Fill your water bottle and leave")
    
    choice2 = input("What do you choose? (1 or 2): ")
    
    if choice2 == "1":
        print("\nThe water gives you magical powers!")
        score += 20
    elif choice2 == "2":
        print("\nYou wisely save the water for later.")
        score += 10
    else:
        print("\nYou ignore the fountain and continue walking.")
        score += 5

# Final Results
print(f"\n--- ADVENTURE COMPLETE ---")
print(f"Congratulations, {player_name}!")
print(f"Your final score: {score}")

if score >= 30:
    print("ğŸ† Excellent! You're a true adventurer!")
elif score >= 20:
    print("ğŸ‘ Good job! You made wise choices.")
elif score >= 10:
    print("ğŸ˜Š Not bad! You survived the adventure.")
else:
    print("ğŸ¤” Better luck next time!")

print("Thanks for playing!")
````

### Lab 3: Simple To-Do List (ê°„ë‹¨í•œ í• ì¼ ëª©ë¡)

**Problem (ë¬¸ì œ)**: Create a simple to-do list manager where users can add tasks, mark them complete, and view their progress.
ì‚¬ìš©ìê°€ ì‘ì—…ì„ ì¶”ê°€í•˜ê³ , ì™„ë£Œë¡œ í‘œì‹œí•˜ê³ , ì§„í–‰ ìƒí™©ì„ ë³¼ ìˆ˜ ìˆëŠ” ê°„ë‹¨í•œ í• ì¼ ëª©ë¡ ê´€ë¦¬ìë¥¼ ë§Œë“œì„¸ìš”.

**Requirements (ìš”êµ¬ì‚¬í•­)**:
- Add new tasks (ìƒˆ ì‘ì—… ì¶”ê°€)
- Mark tasks as complete (ì‘ì—…ì„ ì™„ë£Œë¡œ í‘œì‹œ)
- View all tasks with status (ìƒíƒœì™€ í•¨ê»˜ ëª¨ë“  ì‘ì—… ë³´ê¸°)

**Solution (ì •ë‹µ)**:

````python
# Simple To-Do List Manager
tasks = []
task_id = 1

print("SIMPLE TO-DO LIST MANAGER")
print("=" * 30)

while True:
    print("\n--- MENU ---")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Complete Task")
    print("4. Show Statistics")
    print("5. Exit")
    
    choice = input("Choose option (1-5): ")
    
    if choice == "1":
        # Add task
        print("\n--- ADD TASK ---")
        title = input("Enter task title: ").strip()
        
        if title:
            task = {
                "id": task_id,
                "title": title,
                "completed": False
            }
            tasks.append(task)
            print(f"Added task: {title}")
            task_id += 1
        else:
            print("Task title cannot be empty!")
    
    elif choice == "2":
        # View tasks
        print("\n--- ALL TASKS ---")
        if not tasks:
            print("No tasks found!")
        else:
            print("ID\tStatus\t\tTask")
            print("-" * 35)
            for task in tasks:
                status = "âœ“ Done" if task["completed"] else "â³ Pending"
                print(f"{task['id']}\t{status}\t\t{task['title']}")
    
    elif choice == "3":
        # Complete task
        print("\n--- COMPLETE TASK ---")
        if not tasks:
            print("No tasks found!")
        else:
            # Show pending tasks
            pending_tasks = [task for task in tasks if not task["completed"]]
            if not pending_tasks:
                print("No pending tasks!")
            else:
                print("Pending tasks:")
                for task in pending_tasks:
                    print(f"{task['id']}. {task['title']}")
                
                try:
                    task_id_input = int(input("Enter task ID to complete: "))
                    
                    # Find and complete task
                    task_found = False
                    for task in tasks:
                        if task["id"] == task_id_input:
                            if task["completed"]:
                                print("Task already completed!")
                            else:
                                task["completed"] = True
                                print(f"Completed: {task['title']}")
                            task_found = True
                            break
                    
                    if not task_found:
                        print("Task ID not found!")
                        
                except ValueError:
                    print("Please enter a valid task ID!")
    
    elif choice == "4":
        # Show statistics
        print("\n--- STATISTICS ---")
        if not tasks:
            print("No tasks found!")
        else:
            total_tasks = len(tasks)
            completed_tasks = len([task for task in tasks if task["completed"]])
            pending_tasks = total_tasks - completed_tasks
            completion_rate = (completed_tasks / total_tasks) * 100
            
            print(f"Total tasks: {total_tasks}")
            print(f"Completed: {completed_tasks}")
            print(f"Pending: {pending_tasks}")
            print(f"Completion rate: {completion_rate:.1f}%")
            
            if completion_rate >= 80:
                print("ğŸ‰ Great progress!")
            elif completion_rate >= 50:
                print("ğŸ‘ Keep it up!")
            else:
                print("ğŸ’ª Time to get busy!")
    
    elif choice == "5":
        print("Thanks for using To-Do List Manager!")
        break
    
    else:
        print("Invalid choice! Please try again.")
````
