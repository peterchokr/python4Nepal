### Key Explanation Points

**Syntax Requirements**
- Colon (:) is mandatory after if/elif/else statements
- Indentation is required for code blocks (usually 4 spaces)
- Python is case-sensitive with keywords

**Operators**
- Use == for equality comparison (not =)
- Use != for "not equal to" 
- and: both conditions must be True
- or: at least one condition must be True
- not: reverses the boolean value

**Control Flow**
- elif statements are evaluated in order (top to bottom)
- Once a condition is True, remaining conditions are skipped
- else block executes only if no previous condition was True

**Common Mistakes**
- Using = instead of == for comparison
- Forgetting colons after conditions
- Incorrect indentation causing IndentationError
- Wrong order of elif conditions
