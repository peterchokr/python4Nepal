## 🧮 1. Arithmetic Operators (산술 연산자)

Arithmetic operators perform mathematical calculations, just like a calculator!
산술 연산자는 계산기처럼 수학 계산을 수행합니다!

### 1.1 Basic Arithmetic Operators (기본 산술 연산자)

| Operator (연산자) | Name (이름) | Example (예시) | Result (결과) |
|-------------------|-------------|----------------|---------------|
| `+` | Addition (덧셈) | `5 + 3` | `8` |
| `-` | Subtraction (뺄셈) | `5 - 3` | `2` |
| `*` | Multiplication (곱셈) | `5 * 3` | `15` |
| `/` | Division (나눗셈) | `5 / 3` | `1.6666666666666667` |

#### Simple Examples (간단한 예시)

````
# Shopping calculation (쇼핑 계산)
item_price = 25.99
quantity = 3
total_cost = item_price * quantity
print(f"Total cost: ${total_cost}")  # Output: Total cost: $77.97

# Average calculation (평균 계산)
math_score = 85
english_score = 92
average = (math_score + english_score) / 2
print(f"Average score: {average}")  # Output: Average score: 88.5
````

### 1.2 Advanced Arithmetic Operators (고급 산술 연산자)

| Operator (연산자) | Name (이름) | Example (예시) | Result (결과) |
|-------------------|-------------|----------------|---------------|
| `//` | Floor Division (정수 나눗셈) | `7 // 3` | `2` |
| `%` | Modulo (나머지) | `7 % 3` | `1` |
| `**` | Exponentiation (거듭제곱) | `2 ** 3` | `8` |

#### Understanding Special Operators (특수 연산자 이해하기)

````
# Floor division: How many full groups?
students = 25
groups_of_7 = students // 7
print(f"Number of full groups: {groups_of_7}")  # 3

# Modulo: What's left over?
remaining_students = students % 7
print(f"Students left over: {remaining_students}")  # 4

# Exponentiation: Powers
area = 5 ** 2  # 5 squared
print(f"Area of square: {area}")  # 25

cube_volume = 3 ** 3  # 3 cubed
print(f"Volume of cube: {cube_volume}")  # 27
````

## ⚖️ 2. Comparison Operators (비교 연산자)

Comparison operators compare two values and return `True` or `False`.
비교 연산자는 두 값을 비교하여 `True` 또는 `False`를 반환합니다.

### 2.1 All Comparison Operators (모든 비교 연산자)

| Operator (연산자) | Name (이름) | Example (예시) | Result (결과) |
|-------------------|-------------|----------------|---------------|
| `==` | Equal to (같음) | `5 == 5` | `True` |
| `!=` | Not equal to (같지 않음) | `5 != 3` | `True` |
| `<` | Less than (작음) | `3 < 5` | `True` |
| `>` | Greater than (큼) | `5 > 3` | `True` |
| `<=` | Less than or equal (작거나 같음) | `3 <= 5` | `True` |
| `>=` | Greater than or equal (크거나 같음) | `5 >= 5` | `True` |

### 2.2 Practical Examples (실용적 예시)

````
# Age verification (나이 확인)
user_age = 19
voting_age = 18
driving_age = 16

can_vote = user_age >= voting_age
can_drive = user_age >= driving_age
is_adult = user_age >= 18

print(f"Can vote: {can_vote}")     # True
print(f"Can drive: {can_drive}")   # True
print(f"Is adult: {is_adult}")     # True

# Grade comparison (성적 비교)
alice_score = 85
bob_score = 78
passing_grade = 70

alice_passed = alice_score >= passing_grade
bob_passed = bob_score >= passing_grade
alice_better = alice_score > bob_score

print(f"Alice passed: {alice_passed}")      # True
print(f"Bob passed: {bob_passed}")          # True
print(f"Alice scored higher: {alice_better}") # True
````

## 🧠 3. Logical Operators (논리 연산자)

Logical operators combine multiple conditions to make complex decisions.
논리 연산자는 여러 조건을 결합하여 복잡한 결정을 내립니다.

### 3.1 The Three Logical Operators (3가지 논리 연산자)

| Operator (연산자) | Name (이름) | Description (설명) |
|-------------------|-------------|-------------------|
| `and` | AND | Both conditions must be True (두 조건 모두 참이어야 함) |
| `or` | OR | At least one condition must be True (최소 하나의 조건이 참이어야 함) |
| `not` | NOT | Reverses the condition (조건을 반대로 바꿈) |

### 3.2 Understanding AND Operator (AND 연산자 이해하기)

````
# University admission (대학 입학)
gpa = 3.5
english_score = 85
has_recommendation = True

# ALL requirements must be met (모든 요건이 충족되어야 함)
eligible = (gpa >= 3.0) and (english_score >= 80) and has_recommendation
print(f"Eligible for admission: {eligible}")  # True

# Weather for picnic (피크닉 날씨)
temperature = 22
is_sunny = True
is_raining = False

good_weather = (temperature >= 20) and is_sunny and (not is_raining)
print(f"Good weather for picnic: {good_weather}")  # True
````

### 3.3 Understanding OR Operator (OR 연산자 이해하기)

````
# Payment methods (결제 수단)
has_cash = False
has_credit_card = True
has_mobile_pay = False

# ANY payment method works (어떤 결제 수단이든 작동)
can_pay = has_cash or has_credit_card or has_mobile_pay
print(f"Can make payment: {can_pay}")  # True

# Student discount (학생 할인)
is_student = True
is_senior = False
is_military = False

gets_discount = is_student or is_senior or is_military
print(f"Gets discount: {gets_discount}")  # True
````

### 3.4 Understanding NOT Operator (NOT 연산자 이해하기)

````
# Access control (접근 제어)
is_banned = False
is_verified = True

can_post = is_verified and not is_banned
print(f"Can post messages: {can_post}")  # True

# Office hours (사무실 시간)
is_weekend = False
office_open = not is_weekend
print(f"Office is open: {office_open}")  # True
````

## 📝 4. Assignment Operators (할당 연산자)

Assignment operators provide shortcuts for updating variables.
할당 연산자는 변수를 업데이트하는 단축키를 제공합니다.

### 4.1 Compound Assignment Operators (복합 할당 연산자)

| Operator (연산자) | Equivalent (동등한 표현) | Example (예시) |
|-------------------|-------------------------|----------------|
| `+=` | `x = x + value` | `score += 10` |
| `-=` | `x = x - value` | `score -= 5` |
| `*=` | `x = x * value` | `score *= 2` |
| `/=` | `x = x / value` | `score /= 3` |

### 4.2 Practical Examples (실용적 예시)

````
# Game score system (게임 점수 시스템)
player_score = 100
player_lives = 3

print(f"Initial score: {player_score}")

# Player completes level (레벨 완료)
player_score += 250
print(f"After level: {player_score}")  # 350

# Player loses life (생명 잃음)
player_lives -= 1
print(f"Lives remaining: {player_lives}")  # 2

# Double points bonus (더블 포인트 보너스)
player_score *= 2
print(f"After double points: {player_score}")  # 700

# Bank account (은행 계좌)
balance = 1000.0
print(f"Starting balance: ${balance}")

# Salary deposit (급여 입금)
balance += 3000.0
print(f"After salary: ${balance:.2f}")  # $4000.00

# Rent payment (임대료 지불)
balance -= 800.0
print(f"After rent: ${balance:.2f}")  # $3200.00
````

## 🔢 5. Operator Precedence (연산자 우선순위)

Understanding operator precedence helps you predict the order of operations.
연산자 우선순위를 이해하면 연산 순서를 예측할 수 있습니다.

### 5.1 Order of Operations (연산 순서)

**Remember PEMDAS** (괄호, 지수, 곱셈/나눗셈, 덧셈/뺄셈 순서를 기억하세요):

1. **P**arentheses (괄호) - `( )`
2. **E**xponents (지수) - `**`
3. **M**ultiplication and **D**ivision (곱셈과 나눗셈) - `*`, `/`, `//`, `%`
4. **A**ddition and **S**ubtraction (덧셈과 뺄셈) - `+`, `-`

### 5.2 Examples with Explanation (설명이 있는 예시)

````
# Example 1: Basic precedence
result1 = 10 + 3 * 2
print(f"10 + 3 * 2 = {result1}")  # 16 (not 26!)
# Step by step: 3 * 2 = 6, then 10 + 6 = 16

# Example 2: Using parentheses
result2 = (10 + 3) * 2
print(f"(10 + 3) * 2 = {result2}")  # 26
# Step by step: (10 + 3) = 13, then 13 * 2 = 26

# Example 3: Complex expression
result3 = 2 ** 3 + 4 * 5 - 6 / 2
print(f"2 ** 3 + 4 * 5 - 6 / 2 = {result3}")  # 25.0
# Step by step: 2**3=8, 4*5=20, 6/2=3.0, then 8+20-3.0=25.0

# Use parentheses for clarity (명확성을 위해 괄호 사용)
price = 100
tax_rate = 0.1
total = price + (price * tax_rate)  # Clear what's being calculated
print(f"Total with tax: ${total}")  # $110.0
````

### Lab 1: Calculator Program (계산기 프로그램 만들기)

**Problem (문제)**: Create a calculator program that performs various mathematical operations.
다양한 수학 연산을 수행하는 계산기 프로그램을 만드세요.

**Requirements (요구사항)**:
- Perform all arithmetic operations (+, -, *, /, //, %, **)
- Calculate area of rectangle and circle
- Show clear output with descriptions

**Solution (정답)**:

````
# Simple Calculator Program
print("🧮 CALCULATOR PROGRAM")
print("=" * 30)

# Input numbers (입력 숫자)
num1 = 15
num2 = 4

print(f"Numbers: {num1} and {num2}")
print()

# Basic operations (기본 연산)
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
floor_division = num1 // num2
modulo = num1 % num2
exponentiation = num1 ** num2

print("BASIC OPERATIONS:")
print(f"{num1} + {num2} = {addition}")
print(f"{num1} - {num2} = {subtraction}")
print(f"{num1} × {num2} = {multiplication}")
print(f"{num1} ÷ {num2} = {division:.2f}")
print(f"{num1} // {num2} = {floor_division}")
print(f"{num1} % {num2} = {modulo}")
print(f"{num1} ** {num2} = {exponentiation}")
print()

# Area calculations (넓이 계산)
print("AREA CALCULATIONS:")
length = 12.5
width = 8.0
rectangle_area = length * width
print(f"Rectangle area ({length} × {width}): {rectangle_area}")

radius = 5
pi = 3.14159
circle_area = pi * radius ** 2
print(f"Circle area (radius {radius}): {circle_area:.2f}")
print()

# Practical calculation (실용 계산)
print("FINANCIAL CALCULATION:")
principal = 1000
rate = 0.05
time = 3
simple_interest = principal * rate * time
total = principal + simple_interest
print(f"Investment: ${principal} at {rate*100}% for {time} years")
print(f"Simple interest: ${simple_interest}")
print(f"Total amount: ${total}")
````

### Lab 2: Age Comparison Program (나이 비교 프로그램)

**Problem (문제)**: Create a program that compares ages and determines permissions.
나이를 비교하고 권한을 결정하는 프로그램을 만드세요.

**Requirements (요구사항)**:
- Compare multiple people's ages
- Determine age-based permissions using comparison operators
- Show clear comparison results

**Solution (정답)**:

````
# Age Comparison Program
print("👥 AGE COMPARISON PROGRAM")
print("=" * 35)

# Family members (가족 구성원)
alice_age = 22
bob_age = 19
carol_age = 25
david_age = 16

print("FAMILY AGES:")
print(f"Alice: {alice_age} years old")
print(f"Bob: {bob_age} years old")
print(f"Carol: {carol_age} years old")
print(f"David: {david_age} years old")
print()

# Age comparisons (나이 비교)
print("AGE COMPARISONS:")
print(f"Alice is older than Bob: {alice_age > bob_age}")
print(f"Carol is the oldest: {carol_age > alice_age and carol_age > bob_age and carol_age > david_age}")
print(f"David is the youngest: {david_age < alice_age and david_age < bob_age and david_age < carol_age}")
print(f"Alice and Bob are same age: {alice_age == bob_age}")
print()

# Permission checks (권한 확인)
voting_age = 18
drinking_age = 21
driving_age = 16

print("PERMISSIONS:")
print(f"Alice can vote: {alice_age >= voting_age}")
print(f"Alice can drink: {alice_age >= drinking_age}")
print(f"Bob can vote: {bob_age >= voting_age}")
print(f"Bob can drink: {bob_age >= drinking_age}")
print(f"David can drive: {david_age >= driving_age}")
print(f"David can vote: {david_age >= voting_age}")
print()

# Group statistics (그룹 통계)
total_age = alice_age + bob_age + carol_age + david_age
average_age = total_age / 4
print("GROUP STATISTICS:")
print(f"Total family age: {total_age} years")
print(f"Average age: {average_age:.1f} years")
print(f"All are adults: {alice_age >= 18 and bob_age >= 18 and carol_age >= 18 and david_age >= 18}")
````

### Lab 3: Exam Pass/Fail System (시험 합격 판정 시스템)

**Problem (문제)**: Create an exam grading system using logical operators.
논리 연산자를 사용하여 시험 채점 시스템을 만드세요.

**Requirements (요구사항)**:
- Check multiple subject scores
- Use logical operators for pass/fail criteria
- Provide detailed feedback

**Solution (정답)**:

````
# Exam Grading System
print("🎓 EXAM GRADING SYSTEM")
print("=" * 30)

# Student information (학생 정보)
student_name = "Alex Johnson"
math_score = 85
english_score = 78
science_score = 92
attendance_rate = 95

print(f"Student: {student_name}")
print(f"Math: {math_score}")
print(f"English: {english_score}")
print(f"Science: {science_score}")
print(f"Attendance: {attendance_rate}%")
print()

# Grade boundaries (성적 경계)
excellent_grade = 90
good_grade = 80
pass_grade = 70
min_attendance = 85

# Calculate average (평균 계산)
total_score = math_score + english_score + science_score
average_score = total_score / 3

print("ANALYSIS:")
print(f"Total points: {total_score}")
print(f"Average score: {average_score:.1f}")
print()

# Pass/Fail criteria (합격/불합격 기준)
print("PASS/FAIL CRITERIA:")

# Basic pass: All subjects >= 70 AND attendance >= 85%
basic_pass = (math_score >= pass_grade and 
              english_score >= pass_grade and 
              science_score >= pass_grade and 
              attendance_rate >= min_attendance)
print(f"Basic Pass (all ≥70 + attendance): {basic_pass}")

# Honor roll: Average >= 85 AND at least one subject >= 90
has_excellent_subject = (math_score >= excellent_grade or 
                        english_score >= excellent_grade or 
                        science_score >= excellent_grade)
honor_roll = average_score >= 85 and has_excellent_subject
print(f"Honor Roll (avg ≥85 + one ≥90): {honor_roll}")

# Perfect attendance bonus
perfect_attendance = attendance_rate == 100
print(f"Perfect Attendance: {perfect_attendance}")

# STEM excellence (Math and Science both good)
stem_excellent = math_score >= good_grade and science_score >= good_grade
print(f"STEM Excellence: {stem_excellent}")
print()

# Final determination (최종 판정)
print("FINAL RESULT:")
print(f"Honor Roll qualified: {honor_roll}")
print(f"Basic Pass qualified: {basic_pass}")
print(f"Perfect Attendance: {perfect_attendance}")
print(f"STEM Excellence: {stem_excellent}")

# Areas that need improvement (개선 필요 영역)
print("\nAREAS TO IMPROVE:")
print(f"Math needs improvement: {math_score < pass_grade}")
print(f"English needs improvement: {english_score < pass_grade}")
print(f"Science needs improvement: {science_score < pass_grade}")
print(f"Attendance needs improvement: {attendance_rate < min_attendance}")
````

## 💡 Additional Tips for Success (성공을 위한 추가 팁)

### For Non-Korean Students (한국어가 익숙하지 않은 학생들을 위해)
- **Practice order of operations**: Use PEMDAS rule consistently (연산 순서 연습: PEMDAS 규칙을 일관되게 사용)
- **Test logical expressions**: Try different combinations to understand AND, OR, NOT (논리 표현식 테스트: 다양한 조합으로 AND, OR, NOT 이해)
- **Use parentheses**: When in doubt, add parentheses for clarity (괄호 사용: 확실하지 않을 때는 명확성을 위해 괄호 추가)

### Common Mistakes to Avoid (피해야 할 일반적인 실수)
- **Assignment vs Comparison**: Use `=` for assignment, `==` for comparison (할당과 비교: 할당에는 `=`, 비교에는 `==` 사용)
- **Integer vs Float Division**: Remember `/` gives float, `//` gives integer (정수와 실수 나눗셈: `/`는 실수, `//`는 정수 결과)
- **Logical Operator Order**: AND is evaluated before OR (논리 연산자 순서: AND가 OR보다 먼저 평가)

### Practice Suggestions (연습 제안)
- **Daily calculations**: Use Python for real math problems (일일 계산: 실제 수학 문제에 파이썬 사용)
- **Compare different approaches**: Try both compound and regular assignment (다양한 접근법 비교: 복합 할당과 일반 할당 모두 시도)
- **Test edge cases**: What happens with negative numbers? Zero? (극단적 경우 테스트: 음수나 0은 어떻게 될까?)
